```python
# Adding a new categorical column based on value of a numerical column
conditions = [
    (df['BMI'] < 18.5),
    (df['BMI'] >= 18.5) & (df['BMI'] < 25),
    (df['BMI'] >= 25) & (df['BMI'] < 30),
    (df['BMI'] >= 30)
]

mappings = ['underweight', 'healthy', 'overweight', 'obese']
```

```python
import numpy as np
```

```python
df['bmi_indicator'] = np.select(conditions, mappings, default='unknown')
```

```python
df[['BMI', 'bmi_indicator']].head()
```

```python
df['bmi_indicator'].value_counts().to_frame().reset_index()
```

```python
# Categorizing adult or child
conditions = [
    (df['Age'] < 21),
    (df['Age'] >= 21)
]

mappings = ['Child', 'Adult']
```

```python
df['age_group'] = np.select(conditions, mappings, default='unknown')
```

```python
df[['Age', 'age_group']].head()
```

```python
age_group_df = df['age_group'].value_counts().to_frame().reset_index()
```

```python
age_group_df['percentage'] = (age_group_df['count']/20000) * 100.0
```

```python
age_group_df
```
