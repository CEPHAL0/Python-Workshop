```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_values = np.array([1,2,3,4,5])
y_values = np.array([5,6,7,8,9])
```

```python
# Line Plot -> kind of plot that represents data as points and then 
# joins them using a line which can be used to view the trend

plt.plot(x_values, y_values)
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Sample Line Plot")

plt.show()
```

```python
# Customizing Line Plot

plt.plot(
    x_values, 
    y_values,
    color='green',
    linestyle='--',
    linewidth=2,
    marker='o',
    markersize=15,
    markerfacecolor='red'
)
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Sample Line Plot")

plt.show()
```

```python
df = pd.read_csv("data/lifestyle_final_data.csv")
```

```python
plt.plot(
    df['Weight (kg)'],
    df['lean_mass_kg'],
    color='red',
    linestyle='--',
    linewidth=1,
    marker='o',
    markersize=10,
    markerfacecolor='blue'
)
plt.xlabel("Weight (kg)")
plt.ylabel("Lean Mass Per Kg")
plt.show()
```

```python
plt.scatter(
    np.log(df['prep_time_min']),
    np.log(df['cook_time_min']),
    color='blue',
    s=np.log(df['serving_size_g'])
)
plt.xlabel("Prep Time (in minutes) - Log Scale")
plt.ylabel("Cook Time (in minutes) - Log Scale")

plt.show()
```
