```python
0 - 5 -> Infant
6 - 12 -> Child
13 - 19 -> Teen
20 - 30 -> Baby Adult
31 - 50 -> Adult
51 - 80 -> Old
```

```python
conditions = [
    (df['Age'] >= 0) & (df['Age'] <= 5),
    (df['Age'] >= 6) & (df['Age'] <= 12),
    (df['Age'] >= 13) & (df['Age'] <= 19),
    (df['Age'] >= 20) & (df['Age'] <= 30),
    (df['Age'] >= 31) & (df['Age'] <= 50),
    (df['Age'] >= 51) & (df['Age'] <= 80)
]

mappings = ['Infant', 'Child', 'Teen', 'BabyAdult', 'Adult', 'Old']
```

```python
import numpy as np
df['AgeGroup'] = np.select(conditions, mappings, default='Unknown')
```

```python
df[['Age', 'AgeGroup']]
```

```python
df[['AgeGroup', 'Survived']].value_counts().to_frame().reset_index().sort_values(by='AgeGroup')
```
