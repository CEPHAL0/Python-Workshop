```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Histogram
# Graphical representation of the distribution in the dataset using connected bars to show
# often values fall into specific categories or range

# How is the data distributed?
```

```python
data = np.random.randint(0, 900, size=(1000,))
```

```python
data
```

```python
plt.hist(data)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of a sample data")
plt.show()
```

```python
# Bins 

plt.hist(data, bins=80)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of a sample data")
plt.show()
```

```python
df = pd.read_csv("data/lifestyle_final_data.csv")
df.head()
```

```python
plt.hist(df["Weight (kg)"], bins=20, color='purple')
plt.xlabel("Weight (kg)")
plt.ylabel("Frequency")
plt.title("Distirbution of Weight across the data")
plt.show()
```

```python
plt.hist(
    [df["Max_BPM"], df["Avg_BPM"]], 
    bins=30, 
    stacked=True,
    label=["Max BPM", "Avg BPM"]
)
plt.legend(loc='upper left')
plt.show()
```

```python
plt.hist(
    [df["Max_BPM"], df["Avg_BPM"]], 
    bins=20, 
    stacked=True,
    label=["Max BPM", "Avg BPM"],
    histtype='step'
)
plt.legend(loc='upper left')
plt.show()
```

```python
df.columns
```

```python
plt.hist(
    [df['BMI'], df['BMI_calc']],
    bins=40,
    label=['BMI', 'BMI_Calc'],
    histtype='bar'
)
plt.legend()
plt.title("BMI Distribution")
plt.xlabel("BMI")
plt.ylabel("Frequency")
plt.show()
```

```python
# Box and Whisker Plot
# Median, First Quartile, Third Quartile, Minimum, Maximum and Outliers
```

```python
plt.boxplot(
    df[['Max_BPM', 'Avg_BPM', 'Resting_BPM']],
    tick_labels=['Max BPM', 'Avg BPM', 'Resting BPM'],
    patch_artist=True
)
plt.title("Calories Box Plot")
plt.xlabel("BPM")
plt.show()
```
