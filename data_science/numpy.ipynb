```python
# Universal Functions (ufunc)

# Normal Way without using numpy to add two different arrays
array_1 = [1, 2, 3, 4, 5]
array_2 = [10, 20, 30, 40, 50]

sum_array = []
for index, value in enumerate(array_1):
    sum_value = array_1[index] + array_2[index]
    sum_array.append(sum_value)

sum_array
```

```python
# Add two different arrays using Numpy ufunc

array_1 = np.array([1, 2, 3, 4, 5])
array_2 = np.array([10, 20, 30, 40, 50])
array_3 = np.array([5, 6, 7, 8, 9])

sum_array = np.add(array_1, array_2)

# Add the third array again
sum_array = np.add(sum_array, array_3)


# Subtract two arrays
difference_array = np.subtract(array_2, array_1)

# Multiply two arrays
product_array = np.multiply(array_1, array_2)

# Divide Two arrays
division_array = np.divide(array_2, array_1)

# Remaining functions
# power, mod, remainder, divmod, absolute, 

divmod_array = np.divmod(array_1, array_3)

display(sum_array)
display(difference_array)
display(product_array)
display(division_array)
display(divmod_array)
```

```python
# Rounding Decimals
array_1 = np.array([10.58781, 4.000001])

# Truncate (trunc)
truncated_array = np.trunc(array_1)

# Rounding (round)
# rounding to 0 decimals
rounded_array_0 = np.round(array_1)

# rounding to 1 decimal
rounded_array_1 = np.round(array_1, 1)

# Floor
floor_array = np.floor(array_1)

# Ceiling
ceil_array = np.ceil(array_1)

display(truncated_array)
display(rounded_array_0)
display(rounded_array_1)
display(floor_array)
display(ceil_array)
```

```python
# LCM, GCD

number_1 = 32
number_2 = 64

lcm = np.lcm(number_1, number_2)
gcd = np.gcd(number_1, number_2)

display(lcm)
display(gcd)
```

```python
# Reduce
array_1 = np.array([16, 20, 24, 32, 64, 88])
lcm = np.lcm.reduce(array_1)
gcd = np.gcd.reduce(array_1)
sum_elements = np.add.reduce(array_1)

display(lcm)
display(gcd)
display(sum_elements)
```
