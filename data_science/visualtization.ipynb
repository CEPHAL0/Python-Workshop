```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Make a random data
x = np.arange(10, 15)
y = [100, 200, 300, 250, 400]

x, y
```

```python
plt.bar(x, y)
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.title("Percentage Analysis of Students")
plt.show()
```

```python
fruits = ['apple', 'blueberry', 'cherry' , 'orange']
counts = [100, 350, 45, 200]
bar_labels = ['red', 'blue', '_red', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']

fig, ax = plt.subplots()
ax.bar(fruits, counts, label=bar_labels, color=bar_colors)

ax.set_xlabel("Fruits")
ax.set_title("Fruit Supply by kind and color")
ax.legend(title="Fruit Color")

plt.show()
```

```python
student_names = ['John', 'Alice', 'Matthew', 'James']
student_marks = {
    "Maths": [67, 45, 89, 70],
    "Science": [49, 56, 68, 77],
    "English": [85, 69, 77, 58]
}

x = np.arange(len(student_names))
width = 0.25
multiplier = 0

fig, ax = plt.subplots(layout='constrained')


for subject, marks in student_marks.items():
    offset = width * multiplier

    # plot the bar
    rects = ax.bar(x + offset, marks, width, label=subject)
    ax.bar_label(rects, padding=2)
    multiplier += 1

ax.set_xlabel("Student")
ax.set_ylabel("Subject Marks")
ax.set_xticks(x + width, student_names)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(0, 100)

plt.show()
```
