```python
import pandas as pd
import numpy as np
```

```python
data = {
    "code": ['01620', '01620', '01620', '01620', '01992', '01992', '01992', '01992', '02210', '02210', '02210', '02210'],
    "state": ['MI', 'UT', 'CA', 'TX','MI', 'UT', 'CA', 'TX','MI', 'UT', 'CA', 'TX'],
    "ctc": [33.26, 34.35, 32.50, 33.80, 12.50, 15.30, 11.00, 300.00, 20000.00, 7.00, 10.00, 9.00],
    "tot_srvcs" : [154, 45, 120, 100, 80, 60, 70, 10, 150, 80, 120, 100]
}
```

```python
df = pd.DataFrame(data)
df_copy = df.copy()
```

```python
df
```

```python
df['y'] = np.log(1 + df['ctc'])
```

```python
df
```

```python
code_median = df.groupby('code')['y'].median().to_frame().reset_index()
code_median
```

```python
merged_df = pd.merge(df, code_median, on='code', how='left').rename(columns={'y_x':'y', 'y_y':'c_med'})
```

```python
merged_df
```

```python
merged_df['mad'] = abs(merged_df['y'] - merged_df['c_med'])
merged_df
```

```python
mad_c = merged_df.groupby('code')['mad'].median().to_frame().reset_index()
mad_c
```

```python
merged_df = pd.merge(merged_df, mad_c, on='code', how='left').rename(columns={'mad_x':'mad', 'mad_y':'c_mad'})
merged_df
```

```python
merged_df['z_score'] = 0.6745 * ((merged_df['y']-merged_df['c_med'])/merged_df['c_mad'])
```

```python
merged_df
```

```python
conditions = {
    "filter": [abs(merged_df['z_score']) > 3.5],
    "mappings": "Y"
}

merged_df['flagged'] = np.select(conditions["filter"], conditions["mappings"], default="N")
```

```python
merged_df
```

```python
merged_df['y_upper'] = merged_df['c_med'] + ((3.5 * merged_df['c_mad'])/0.6745)
merged_df['y_lower'] = merged_df['c_med'] - ((3.5 * merged_df['c_mad'])/0.6745)
```

```python
merged_df
merged_df_copy = merged_df.copy()

merged_df
```

```python
merged_df.loc[(merged_df['flagged'] == 'Y') & (merged_df['y'] > merged_df['y_upper']), 'final_y'] = merged_df['y_upper']
merged_df.loc[(merged_df['flagged'] == 'Y') & (merged_df['y'] < merged_df['y_lower']), 'final_y'] = merged_df['y_lower']
merged_df.loc[merged_df['flagged'] == 'N', 'final_y'] = merged_df['y']
```

```python
merged_df
```

```python
merged_df['final_ctc'] = np.exp(merged_df['final_y']) - 1
merged_df
```
