```python
# Pie Chart
# Pie Chart is the kind of chart that shows the proportion in the data
# For example: Plotting the iOS, Android and Windows market share in mobile smartphones
```

```python
import matplotlib.pyplot as plt
```

```python
market_shares = [45, 50, 5]
labels = ['iOS', 'Android', 'Windows']
```

```python
plt.pie(market_shares, labels=labels)
plt.title("Market Share of various OS in Smartphones")
plt.show()
```

```python
colors = ['brown', 'green', 'purple', 'black']
plt.pie(market_shares, labels=labels, autopct='%1.2f%%', colors=colors, startangle=180, shadow=True)
plt.show()
```

```python
explode = (0, 0, 0.2)
plt.pie(market_shares, labels=labels, explode=explode)
plt.show()
```

```python
hatches = ['o', '\\', '*']
plt.pie(market_shares, labels=labels, hatch=hatches)
plt.legend(loc='lower right')
plt.show()
```

```python
import pandas as pd
```

```python
df = pd.read_csv("data/Titanic-Dataset.csv")
```

```python
df.head()
```

```python
import numpy as np
survived_value = df['Survived'].value_counts().to_frame().reset_index()
survived_value

mappings = [
    survived_value['Survived'] == 0,
    survived_value['Survived'] == 1
]

values = ['Didnt survive', 'Survive']

survived_value['Survived_Name'] = np.select(mappings, values, default='Unknown')
survived_value
```

```python
plt.pie(survived_value['count'], labels=survived_value['Survived_Name'], autopct='%1.1f%%', explode=[0, 0.05])
plt.legend(loc='lower right', ncols=2)
plt.title("Survived and Not Survived Counts")
plt.show()
```

```python
gender = df['Sex'].value_counts().to_frame().reset_index()

plt.pie(
    gender['count'], 
    labels=gender['Sex'], 
    colors=['black', 'pink'], 
    autopct='%1.1f%%', 
    textprops={'color':'white'},
    explode=[0, 0.025]
)

plt.legend(loc='lower right', ncols=2)
plt.title("Gender who were in Titanic Ship")

plt.show()
```
