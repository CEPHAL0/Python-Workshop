```python
def simple_rnn_forward(x, hidden_dims=5):
    Wxh = np.random.randn(hidden_dims, 1)
    Whh = np.random.randn(hidden_dims, hidden_dims)
    Why = np.random.randn(1, hidden_dims)


    h = np.zeros((hidden_dims, 1))

    for t in range(x.shape[0]):
        h = np.tanh(Wxh @ x[t].reshape(1,1) + Whh @ h)

    y_pred = Why @ h
    return y_pred

data = np.array([0.1, 0.2, 0.3])
print(simple_rnn_forward(data))
```

```python
import gensim
from gensim.models import Word2Vec
```

```python
sentences = [
    ["i", "love", "machine", "learning"],
    ["word2vec", "creates", "word", "embeddings"],
    ["king", "queen", "man", "woman"]
]


model = Word2Vec(
    sentences,
    vector_size=100,
    window=3,
    min_count=1,
    workers=1,
    sg=1
)
```

```python
print(model.wv["king"])
```

```python
model.wv.similarity("king", "queen")
```

```python
model.wv.most_similar("king")
```
